## Pros/Cons

### Pros

- визуально код проще
- мало один кода 
- сложные программы пишутся проще
- лучшая организация и структура кода

### Cons

- higher learning curve vs procedural programming
- overhead: необходимо хранить информацию о типах и иерархии
- может привести к излишней сложности для простых задач
- сложнее отладка при глубокой иерархии классов
- производительность может быть ниже чем procedural код

## Области класса (Access Modifiers)

### public
- члены класса доступны из любого места программы
- часть публичного интерфейса класса
- используется для методов и данных, предназначенных для внешнего использования

### protected
- члены доступны только внутри класса и его наследников (потомков)
- не доступны из внешнего кода
- используется для внутреннего API между базовым и производными классами

### private
- члены доступны только внутри класса
- полностью скрыты от внешнего кода и наследников
- максимальная инкапсуляция

## Реализация ООП в разных языках

### C++
- явное объявление прав доступа (public, protected, private)
- поддержка множественного наследования
- конструкторы и деструкторы
- виртуальные методы для полиморфизма
- шаблоны (templates) для обобщенного программирования

#### Особенности C++:

**this pointer**
- Указатель на текущий объект внутри метода класса
- Используется для доступа к членам объекта или их различия от параметров
- Неявно передается во все методы не-static

**Smart Pointers**
- `unique_ptr` — исключительное владение ресурсом (не копируется)
- `shared_ptr` — совместное владение через reference counting
- Автоматическое управление памятью (RAII паттерн)
- Предпочтительнее raw pointers для безопасности

**Другие особенности**
- Ручное управление памятью (new/delete) при использовании raw pointers
- Конструкторы копирования и move семантика
- Operator overloading
- Множественное наследование (может привести к diamond problem)

### Java
- все классы наследуют от Object
- поддержка только единственного наследования (single inheritance)
- интерфейсы для множественного наследования интерфейсов
- автоматическая сборка мусора (garbage collection)
- все методы по умолчанию виртуальные

#### Особенности Java:

**Reference Types**
- Все объекты в Java — это references (существуют только в heap)
- примитивные типы (int, boolean и т.д.) хранятся в stack
- null — допустимое значение для всех reference переменных

**Garbage Collection**
- Автоматическое удаление неиспользуемых объектов
- Разработчик не должен вызывать delete или free
- Может привести к непредсказуемым паузам в выполнении

**Другие особенности**
- this — ссылка (не указатель) на текущий объект
- final keyword для предотвращения переопределения методов/наследования
- Иерархия исключений (try-catch-finally)
- Reflection API для динамического анализа классов

### Python
- динамическая типизация
- множественное наследование поддерживается
- методы не требуют явного объявления (duck typing)
- все данные по умолчанию публичные (convention: _ для protected, __ для private)
- декораторы (@property, @staticmethod и т.д.)

## Примеры

Простой пример класса для дальнейшего толкования концепций.

### C++ пример

Базовый класс Animal и наследуемый класс Dog.

```python
#include <iostream>
#include <memory>
using namespace std;

// Базовый класс
class Animal {
protected:
    string name;
public:
    Animal(string n) : name(n) {}
    virtual void speak() {
        cout << "Animal sound" << endl;
    }
    virtual ~Animal() {}
};

// Производный класс
class Dog : public Animal {
private:
    int age;
public:
    Dog(string n, int a) : Animal(n), age(a) {}
    void speak() override {
        cout << name << " says: Woof!" << endl;
    }
    
    // this pointer пример
    void printInfo() {
        cout << "Address of this: " << this << endl;
        cout << this->name << " is " << this->age << " years old" << endl;
    }
};

int main() {
    // Raw pointer (опасно)
    // Dog* dog = new Dog("Rex", 5);
    // delete dog;  // нужно вручную удалить
    
    // Smart pointer (безопасно)
    unique_ptr<Dog> dog1 = make_unique<Dog>("Rex", 5);
    dog1->speak();
    dog1->printInfo();
    // автоматически удалится при выходе из scope
    
    // Shared pointer (reference counting)
    shared_ptr<Dog> dog2 = make_shared<Dog>("Buddy", 3);
    shared_ptr<Dog> dog3 = dog2;  // оба указывают на один объект
    cout << "Reference count: " << dog2.use_count() << endl;
    
    return 0;
}
```

### Java пример

Базовый класс Vehicle и наследуемый класс Car.

```python
// Базовый класс
public class Vehicle {
    protected String name;
    
    public Vehicle(String name) {
        this.name = name;
    }
    
    public void drive() {
        System.out.println("Vehicle is driving");
    }
}

// Производный класс
public class Car extends Vehicle {
    private int doors;
    
    public Car(String name, int doors) {
        super(name);
        this.doors = doors;
    }
    
    @Override
    public void drive() {
        System.out.println(name + " car with " + doors + " doors is driving");
    }
    
    // this пример
    public void printInfo() {
        System.out.println("Object: " + this);
        System.out.println(this.name + " has " + this.doors + " doors");
    }
    
    public int getDoors() {
        return this.doors;  // this явно (хотя можно this опустить)
    }
    
    // final метод — не может быть переопределен
    public final void honk() {
        System.out.println("Beep!");
    }
    
    public static void main(String[] args) {
        Car car1 = new Car("Tesla", 4);
        car1.drive();
        car1.printInfo();
        
        Car car2 = car1;  // обе переменные ссылаются на один объект
        System.out.println("car1 == car2: " + (car1 == car2));
        
        // После выхода из scope объект будет удален garbage collector'ом
        // (если на него нет других ссылок)
    }
}
```

### Python пример

Базовый класс Shape и наследуемый класс Circle.

```python
import math

# Базовый класс
class Shape:
    def __init__(self, name):
        self._name = name  # protected (по соглашению)
    
    def area(self):
        raise NotImplementedError("Subclass must implement area()")
    
    def __str__(self):
        return f"Shape: {self._name}"

# Производный класс
class Circle(Shape):
    def __init__(self, name, radius):
        super().__init__(name)
        self.__radius = radius  # private (name mangling)
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def get_radius(self):
        return self.__radius

# Использование
circle = Circle("MyCircle", 5)
print(circle)
print(f"Area: {circle.area():.2f}")
```
